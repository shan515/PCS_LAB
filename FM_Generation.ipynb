FM Generation 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#input the required info 
Amp_c = float(input('Enter carrier amplitude: '))
Amp_m = float(input('Enter modulating signal amplitude: '))
freq_c = float(input('Enter carrier frequency : '))
freq_m = float(input('Enter modulating signal frequency: '))
deviation_ratio = float(input('Enter deviation_ratio: '))
```

```python
t = np.linspace(0,1,4000)        
carrier = Amp_c * np.cos(2*np.pi*freq_c*t)
modulator = Amp_m * np.cos(2*np.pi*freq_m*t)
resultant = np.cos(2*np.pi*freq_c*t + deviation_ratio*modulator)
```

```python
plt.subplot(3,1,1)
plt.title('FM Generation')
plt.plot(modulator,color='green')
plt.ylabel('Amplitude')
plt.xlabel('Message signal')

plt.subplot(3,1,2)
plt.plot(carrier, color='red')
plt.ylabel('Amplitude')
plt.xlabel('Carrier signal')

plt.subplot(3,1,3)
plt.plot(resultant, color="blue")
plt.ylabel('Amplitude')
plt.xlabel('FM signal')

plt.subplots_adjust(hspace=1)
plt.rc('font', size=15)
fig = plt.gcf()
fig.set_size_inches(16, 9)
```
