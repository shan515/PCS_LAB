AM Generation   


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#input from user
Amp_c = float(input('Enter carrier amplitude: '))
#Amp_m = float(input('Enter modulating signal amplitude: '))
freq_c = float(input('Enter carrier frequency : '))
#freq_m = float(input('Enter modulating signal frequency: '))
modulation_index = float(input('Enter modulation index: '))
msg_func = input("Enter message signal equation: ")
```

```python
t = np.linspace(-1,1,4000)        
carrier = Amp_c * np.cos(2*np.pi*freq_c*t)
m_p = modulation_index * Amp_c
msg_sig = m_p*eval("np."+msg_func) 
msg_ = eval("np."+msg_func)
resultant = (Amp_c + msg_sig)* carrier
```

```python
#plotting part
plt.subplot(3,1,1)
plt.title('AM Generation')
plt.plot(msg_,color='green')
plt.ylabel('Amplitude')
plt.xlabel('Message signal')

plt.subplot(3,1,2)
plt.plot(carrier, color='red')
plt.ylabel('Amplitude')
plt.xlabel('Carrier signal')

plt.subplot(3,1,3)
plt.plot(resultant, color="blue")
plt.ylabel('Amplitude')
plt.xlabel('AM signal')

plt.subplots_adjust(hspace=1)
plt.rc('font', size=15)
fig = plt.gcf()
fig.set_size_inches(16, 9)
```
